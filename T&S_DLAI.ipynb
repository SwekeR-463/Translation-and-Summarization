```python
#to suppress warning messages
from transformers.utils import logging
logging.set_verbosity_error()
```

```python
#translation pipeline
from transformers import pipeline
import torch
```

```python
translator = pipeline("translation",
                      model="facebook/nllb-200-distilled-600M",
                      torch_dtype=torch.bfloat16)
```

```python
text = """\
Our Earth is diverse.\
The Sun is radiant.\
The Moon is mysterious.\
The asteroid belt is scattered."""
```

```python
text_translated = translator(text,
                             src_lang="eng_Latn",
                             tgt_lang="hin_Deva")
```

```python
print(text_translated)
```

```python
#in order to run the rest of code freeing up memory on the machine
import gc
del translator
gc.collect()
```

```python
#summarization pipeline
summarizer = pipeline("summarization",
                      model="facebook/bart-large-cnn",
                      torch_dtype=torch.bfloat16)
```

```python
text = """Bhubaneswar, the capital of Odisha,
          is a bustling metropolis steeped in history and culture.
          With a population exceeding 1.5 million spread over approximately 422 square kilometers,
          it serves as the administrative, economic, and cultural nucleus of the state.
          Boasting ancient temples alongside modern infrastructure,
          Bhubaneswar embodies the fusion of tradition and progress,
          attracting visitors and residents alike with its rich heritage and dynamic urban landscape."""
```

```python
summary = summarizer(text,
                     min_length = 10,
                     max_length = 100)
```

```python
print(summary)
```
